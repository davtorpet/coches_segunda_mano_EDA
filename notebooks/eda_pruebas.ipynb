```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import pearsonr
from scipy.stats import chi2_contingency
from scipy.stats import mannwhitneyu
from scipy import stats
```

```python
df_eda = pd.read_csv("dffinal.csv")
df_eda.head(2)
```

```python
df_eda.drop(columns=["dealer_country_code","dealer_city"],inplace=True)
df_eda.head(5)
```

```python
df_eda["Car and Model"] = df_eda["make"] + " " + df_eda["model"]
df_eda.head(2)
```

```python
bins = [0, 7000, 20000, 50000, 80000, float('inf')]
labels = ["Bajo", "Moderado", "Alto", "Muy Alto", "Excesivo"]

df_eda["precio_categorias"] = pd.cut(df_eda["price"], bins=bins, labels=labels, right=True)
```

```python
bins = [-1,5000,20000,50000,100000,200000,float("inf")]
labels = ["Muy Poco","Poco","Moderado","Medio Alto","Alto","Muy Alto"]
df_eda["kms_categoria"] = pd.cut(df_eda["kms"],bins=bins,labels=labels,right=True)
```

```python
df_eda["years_old"] = 2025 - df_eda["year"]
df_eda.head(3)
```

```python
df_eda.info()
```

```python
locations = df_eda.loc[df_eda["state"]=="Unknown","location"].unique()
locations
```

```python
#Vamos a utilizar la API de openstreetmap para conseguir las comunidades autonomas , la latitud y la longitud
import requests
import time

def conseguir_state_lan_long(location):
    try:
        time.sleep(1.2)
        url= f"https://nominatim.openstreetmap.org/search?format=json&q={location},Espa%C3%B1a"
        respuesta = requests.get(url,headers={"User-Agent":"Mozilla/5.0"})
        dato = respuesta.json()

        if dato:
            estado = dato[0].get("display_name","No existe").split(",")[-2].strip()
            latitud = dato[0].get("lat")
            longitud = dato[0].get("lon")
            return [location,latitud,longitud,estado]
        else:
            print("No es posible, algo falla en la respuesta")
    except:
        print(f"Error en {location}")
```

```python
lista = []
for location in locations:
    lista.append(conseguir_state_lan_long(location=location))
```

```python
for item in lista:
    if item is None:
        continue
    
    location = item[0]
    lat, lon, state = item[1], item[2], item[3]
    
    condition = (df_eda["location"] == location) & (df_eda["lat"].isna())
    df_eda.loc[condition, ["lat", "long", "state"]] = float(lat), float(lon), state
```

```python
df_eda.info()
```

```python
#Vamos a quitar esas filas que no tienen, ya que no nos interesan tanto
df_eda = df_eda.loc[~((df_eda["lat"].isna()) & (df_eda["long"].isna()))]
```
